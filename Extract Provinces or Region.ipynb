# Spatial Join (Shape file and model dataset)

```python
import dask.dataframe as dd

ddf = dd.read_csv(
    "/afs/enea.it/por/user/nafis/PFS/por/Thesis/dataset/csv/*",
    parse_dates=['time']
)
```

```python
print(ddf.dtypes)
```

```python
import dask_geopandas

ddf["x"] = ddf["x"] * 1000
ddf["y"] = ddf["y"] * 1000

ddf = dask_geopandas.from_dask_dataframe(
    ddf,
    geometry=dask_geopandas.points_from_xy(ddf, "x", "y", z=None),
)

ddf.head(3)
```

```python
import dask_geopandas



ddf = dask_geopandas.from_dask_dataframe(
    ddf,
    geometry=dask_geopandas.points_from_xy(ddf, "x", "y", z=None),
)

ddf.head(3)
```

```python
ddf = ddf.set_crs(32632)
ddf.head(3)
```

```python
# Perform the spatial join

italy_with_model_data_gdf1 = ddf.sjoin(italy_gdf, predicate="within")
italy_with_model_data_gdf2 = ddf.sjoin(italy_prov_gdf, predicate="within")

# View the resulting GeoDataFrame
italy_with_model_data_gdf2.head()
```

```python
print(italy_with_model_data_gdf2.dtypes)
```
